```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report


# Load data and split into feature and target variables
data_loaded = load_breast_cancer()
df = pd.DataFrame(data_loaded.data, columns=data_loaded.feature_names)
target = data_loaded.target[[10,50,85]]

print(target)

# View first 5 rows of the data
print(df.head(5))

# How frequently does the positive target occur?
# let's get the occurrences of the positive target
positive_number = np.sum(data_loaded.target > 0)
total_count = len(data_loaded.target)
frequency_positive = positive_number / total_count

print(f"Positive Target Count: {positive_number}")
print(f"Frequency of Positive Target: {frequency_positive:.2%}")


# Generate summary statistics for the data
print(df.shape)
print(df.describe())

# Create a pairplot for the first few features
selected_features = df.loc[:, df.columns[:10]]
selected_features['target'] = data_loaded.target
sns.pairplot(selected_features, hue='target', diag_kind='kde', corner=True)
plt.show()

#Create a correlation coefficeint heatmap

#let's create the matrix first
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Coefficient Heatmap")
plt.show()

# Create a boxplot for mean radius by target type
plt.figure(figsize=(8, 6))
sns.boxplot(x=data_loaded.target, y=df['mean radius'])
plt.title("Boxplot of Mean Radius by Target Type")
plt.xlabel("Target Type (0 = Malignant, 1 = Benign)")
plt.ylabel("Mean Radius")
plt.show()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    df,
    data_loaded.target,
    test_size=0.2,
    random_state=47
)

print(f"Training Set Size: {X_train.shape[0]} samples")
print(f"Test Set Size: {X_test.shape[0]} samples")

# Build and train logistic regression model
# the logistic regression model
logistic_model = LogisticRegression(max_iter=10000, random_state=47)

# Train the model
logistic_model.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = logistic_model.score(X_test, y_test)

# Generate a confusion matrix
#let's create the classification report first
y_pred = logistic_model.predict(X_test)

# Generate the classification report
class_report = classification_report(y_test, y_pred, target_names=data_loaded.target_names)

# Extract coefficients and intercept
coefficients = logistic_model.coef_[0]

# Calculate standard deviations of the features
feature_std = df.std()

# Normalize coefficients by dividing by standard deviation
normalized_coefficients = coefficients / feature_std

# Create a DataFrame with features and coefficients
feature_names = df.columns
coeff_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Add a column for absolute value of coefficients
coeff_df['Abs Coefficient'] = coeff_df['Coefficient'].abs()

# Sort by absolute value of coefficients in descending order
sorted_coeff_df = coeff_df.sort_values(by='Abs Coefficient', ascending=False)

# Print the sorted DataFrame
print(sorted_coeff_df)

importance_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
importance_df['Absolute Coefficient'] = importance_df['Coefficient'].abs()

# Sort by absolute coefficient value
importance_df = importance_df.sort_values(by='Absolute Coefficient', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Absolute Coefficient'], color='skyblue')
plt.xlabel("Absolute Coefficient Value")
plt.ylabel("Feature")
plt.title("Feature Importances Based on Logistic Regression Coefficients")
plt.show()

```
